```python
#import folders and dependencies
import pandas as pd
from pathlib import Path
import numpy as np
```

```python
#Variables needed later on, set to initial values
total_schools = 0
total_students = 0
total_budget = 0
average_math_score = 0
average_reading_score = 0
number_passing_math = 0
number_passing_reading = 0
number_both_passing = 0
percent_passing_math = 0
percent_passing_reading = 0
percent_both_passing = 0
```

```python
#import files to read 
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

#Read data into Panda DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

#Combine data in one dataset
complete_student_school_data = pd.merge(student_data, school_data, how = "left", on = ["school_name", "school_name"])
complete_student_school_data.head()
```

```python
#District Summary

total_schools = complete_student_school_data["school_name"].nunique()
#total number of schools

total_students = complete_student_school_data["Student ID"].nunique()
#total number of students

total_budget = school_data["budget"].sum()
#total budget

average_math_score = complete_student_school_data["math_score"].mean()
#average math score

average_reading_score = complete_student_school_data["reading_score"].mean()
#average reading score

number_passing_math = len(complete_student_school_data[complete_student_school_data["math_score"] >= 70])
percent_passing_math = (number_passing_math/total_students) * 100
#percent passing math

number_passing_reading = len(complete_student_school_data[complete_student_school_data["reading_score"] >= 70])
percent_passing_reading = (number_passing_reading/total_students) * 100
#percent passing reading

number_both_passing = complete_student_school_data[(complete_student_school_data["math_score"] >= 70) & (complete_student_school_data["reading_score"] >= 70)].count()["Student ID"]
percent_both_passing = (number_both_passing/total_students) * 100
#percent passing reading AND reading

#print(total_schools)
#print(total_students)
#print(total_budget)
#print(average_math_score)
#print(average_reading_score)
#print(percent_passing_math)
#print(percent_passing_reading)
#print(percent_both_passing)

#above was added to see all outputs at once before making dataframe, was not important 
```

```python
#create dataframe to hold above
district_summary_dataframe = pd.DataFrame({
    "Total Schools" : [total_schools],
    "Total Students" : [total_students],
    "Total Budget" : [total_budget],
    "Average Math Score" : [average_math_score],
    "Average Reading Score" : [average_reading_score],
    "Percent Passing Math" : [percent_passing_math],
    "Percent Passing Reading" : [percent_passing_reading],
    "Percent Passing Math and Reading" : [percent_both_passing]
})

district_summary_dataframe [["Total Schools",
                             "Total Students",
                             "Total Budget",
                             "Average Math Score",
                             "Average Reading Score",
                             "Percent Passing Math",
                             "Percent Passing Reading",
                             "Percent Passing Math and Reading"]]
                             
```

```python
#School Summary
#Top 5 best performing schools

grouped_school_dataframe = complete_student_school_data.groupby(["school_name"])
#school name

school_type = grouped_school_dataframe["type"].first()
#school type

total_students_school_summary = grouped_school_dataframe["Student ID"].count()
#total students in top 5 schools

total_school_budget_school_summary = grouped_school_dataframe["budget"].first()
#total budget in top 5 schools

per_student_budget_school_summary = total_school_budget_school_summary/total_students_school_summary
#per student budget in top 5 schools

average_math_score_school_summary = grouped_school_dataframe["math_score"].mean()
#average math score in top 5 schools

average_reading_score_school_summary = grouped_school_dataframe["reading_score"].mean()
#average reading score in top 5 schools

passing_math_school_summary = complete_student_school_data[complete_student_school_data["math_score"] >= 70].groupby(["school_name"])["math_score"].count()
percent_passing_math_school_summary = (passing_math_school_summary/total_students_school_summary) * 100
#percent passing math in top 5 schools

passing_reading_school_summary = complete_student_school_data[complete_student_school_data["reading_score"] >= 70].groupby(["school_name"])["reading_score"].count()
percent_passing_reading_school_summary = (passing_reading_school_summary/total_students_school_summary) * 100
#percent passing reading in top 5 schools

passing_both_school_summary = complete_student_school_data[
    (complete_student_school_data["reading_score"] >= 70) & (complete_student_school_data["math_score"] >= 70)
]
passing_both_school_summary = passing_both_school_summary.groupby(["school_name"]).size()
percent_passing_both_school_summary = (passing_both_school_summary/total_students_school_summary) *100
```

```python
#dataframe to hold above
school_summary_dataframe = pd.DataFrame({"School Type" : school_type,
                                         "Total Students" : total_students_school_summary,
                                         "Total Budget" : total_school_budget_school_summary,
                                         "Per Student Budget" : per_student_budget_school_summary,
                                         "Average Math Score" : average_math_score_school_summary,
                                         "Average Reading Score" : average_reading_score_school_summary,
                                         "Percent Passing Math" : percent_passing_math_school_summary,
                                         "Percent Passing Reading" : percent_passing_reading_school_summary,
                                         "Percent Passing Both" : percent_passing_both_school_summary
                                        })
#this dataframe holds data grouped by school name and type
```

```python
school_summary_dataframe = school_summary_dataframe.sort_values(["Percent Passing Both"], ascending = False)
school_summary_dataframe [["School Type",
                             "Total Students",
                             "Total Budget",
                             "Per Student Budget",
                             "Average Math Score",
                             "Average Reading Score",
                             "Percent Passing Math",
                             "Percent Passing Reading",
                             "Percent Passing Both"]].head()
 #this displays the top 5 schools                        
```

```python
school_summary_dataframe = school_summary_dataframe.sort_values(["Percent Passing Both"], ascending = True)
school_summary_dataframe [["School Type",
                             "Total Students",
                             "Total Budget",
                             "Per Student Budget",
                             "Average Math Score",
                             "Average Reading Score",
                             "Percent Passing Math",
                             "Percent Passing Reading",
                             "Percent Passing Both"]].head()
#this displays the botttom 5 schools
```

```python
#math scores by grade
grade9_math_scores = complete_student_school_data.loc[complete_student_school_data["grade"] == "9th"].groupby(["school_name"])["math_score"].mean()
grade10_math_scores = complete_student_school_data.loc[complete_student_school_data["grade"] == "10th"].groupby(["school_name"])["math_score"].mean()
grade11_math_scores = complete_student_school_data.loc[complete_student_school_data["grade"] == "11th"].groupby(["school_name"])["math_score"].mean()
grade12_math_scores = complete_student_school_data.loc[complete_student_school_data["grade"] == "12th"].groupby(["school_name"])["math_score"].mean()

#combine above into dataframe
grade_math_summary = pd.DataFrame({"9th" : grade9_math_scores,
                                   "10th" : grade10_math_scores,
                                   "11th" : grade11_math_scores,
                                   "12th" : grade12_math_scores})
grade_math_summary [["9th", "10th", "11th", "12th"]]
```

```python
#reading scores by grade
grade9_reading_scores = complete_student_school_data.loc[complete_student_school_data["grade"] == "9th"].groupby(["school_name"])["reading_score"].mean()
grade10_reading_scores = complete_student_school_data.loc[complete_student_school_data["grade"] == "10th"].groupby(["school_name"])["reading_score"].mean()
grade11_reading_scores = complete_student_school_data.loc[complete_student_school_data["grade"] == "11th"].groupby(["school_name"])["reading_score"].mean()
grade12_reading_scores = complete_student_school_data.loc[complete_student_school_data["grade"] == "12th"].groupby(["school_name"])["reading_score"].mean()

#combine above into dataframe
grade_reading_summary = pd.DataFrame({"9th" : grade9_reading_scores,
                                      "10th" : grade10_reading_scores,
                                      "11th" : grade11_reading_scores,
                                      "12th" : grade12_reading_scores})
grade_reading_summary [["9th", "10th", "11th", "12th"]]
```
